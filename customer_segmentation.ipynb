```python
df=df.drop(["Gender","CustomerID"],axis=1)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Assuming X is your feature matrix
inertia = []
k_range = range(1, 11)  # Testing k from 1 to 10
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42).fit(df)
    inertia.append(kmeans.inertia_)

plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python

from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in k_range:
    if k > 1: # Start from k=2 to avoid the error
        kmeans = KMeans(n_clusters=k, random_state=0).fit(df)
        labels = kmeans.labels_
        silhouette_scores.append(silhouette_score(df, labels))

plt.plot(range(2, 11), silhouette_scores, marker='o') # Adjust the x-axis range
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.show()
```

```python

from sklearn.decomposition import PCA

# Fit K-means and get cluster labels
kmeans = KMeans(n_clusters=5, random_state=0).fit(df)
labels = kmeans.labels_

# Dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(df)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D PCA of K-means Clusters')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
df["labels"]=labels
```
